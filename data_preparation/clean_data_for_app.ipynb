```python
#Removing numbers
pro.loc[:,"estu_nucleo_pregrado"] = pro.estu_nucleo_pregrado.apply(lambda x : re.sub('[0123456789]','',x))
```

```python
#Remove trailing spaces
#Trim function
trim_function = lambda x : re.findall("^\s*(.*?)\s*$",str(x))[0]

pro.loc[:,"estu_prgm_academico"] = pro.estu_prgm_academico.apply(trim_function)
pro.loc[:,"estu_nucleo_pregrado"] = pro.estu_nucleo_pregrado.apply(trim_function)
```

```python
#Removing accents

#Remove tilde function
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
        ("ü", "u"),
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s

#Removing tilde
pro.loc[:,"estu_prgm_academico"] = pro.estu_prgm_academico.apply(normalize)
pro.loc[:,"estu_inst_departamento"] = pro.estu_inst_departamento.apply(normalize)
pro.loc[:,"estu_inst_municipio"] = pro.estu_inst_municipio.apply(normalize)
pro.loc[:,"estu_nucleo_pregrado"] = pro.estu_nucleo_pregrado.apply(normalize)
```

```python
#Fix municipio
pro.loc[:,"estu_inst_municipio"] = pro.estu_inst_municipio.apply(lambda x : re.sub('BOGOTA D.C.','BOGOTA, D.C.',x))
```
